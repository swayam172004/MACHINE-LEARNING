```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("exoplanets2.csv")
df.head(3)
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(25,10))
sns.boxplot(x='sy_pnum',data=df)
plt.plot()
```

```python
sns.distplot(df['sy_pnum'])
plt.plot()
```

```python
l = [1,5,2,4,3,2,1,5,0,4,2,100]
np.mean(l)
```

```python
q1 = df['sy_pnum'].quantile(0.25)
q3 = df['sy_pnum'].quantile(0.75)
```

```python
IQR = q3 -q1
```

```python
min_range = q1 - IQR
max_range = q3 + (1.5*IQR)
max_range
```

```python
new_data = df[df['sy_pnum']<=max_range]
```

```python
plt.figure(figsize=(25,10))
sns.boxplot(x='sy_pnum',data=new_data)
plt.plot()
```

# z-score

# z = (x-u)/s.d

```python
sns.boxplot(x='sy_pnum',data=df)
plt.plot()
```

```python
sns.distplot(df['sy_pnum'])
plt.plot()
```

```python
min_range = df['sy_pnum'].mean()-(3*df['sy_pnum'].std())
max_range = df['sy_pnum'].mean()+(3*df['sy_pnum'].std())
min_range
```

```python
max_range
```

```python
new_d = df[df['sy_pnum']<=max_range]
new_d.shape
```

```python
sns.boxplot(x='sy_pnum',data=new_d)
plt.plot()
```

```python
z_score = (df['sy_pnum'] - df['sy_pnum'].mean())/(df['sy_pnum'].std())
```

```python
df['z_score'] = z_score
```

```python
df[df['z_score']<3]
```
